# Chamando flows do Langflow via API

Este notebook demonstra como carregar a API key gerada automaticamente e chamar um flow do Langflow.

## 1. Carregar credenciais

```python
from dotenv import load_dotenv
import os

# /config é o volume compartilhado com o serviço langflow-init
load_dotenv("/config/langflow.env")

LANGFLOW_URL = os.environ["LANGFLOW_URL"]
API_KEY      = os.environ["LANGFLOW_API_KEY"]

print(f"LANGFLOW_URL : {LANGFLOW_URL}")
print(f"LANGFLOW_API_KEY : {API_KEY[:12]}{'*' * 20}")
```

## 2. Listar flows disponíveis

```python
import requests

headers = {"x-api-key": API_KEY}

resp = requests.get(f"{LANGFLOW_URL}/api/v1/flows/", headers=headers)
resp.raise_for_status()

flows = resp.json()
for flow in flows:
    print(f"{flow['id']}  →  {flow['name']}")
```

## 3. Chamar um flow

Substitua `FLOW_ID` pelo ID do flow listado acima.

```python
FLOW_ID = "fa338b94-389d-46c3-a5b5-d7437d8d86b6"

input_json = '{"name": "Eduardo", "age": 34, "email": "edmesmor@bol.com.br"}'

payload = {
    "input_value": input_json,
    "input_type": "text",
    "output_type": "text",
}

resp = requests.post(
    f"{LANGFLOW_URL}/api/v1/run/{FLOW_ID}",
    json=payload,
    headers=headers,
)
resp.raise_for_status()
```

## 4. Extrair só o texto da resposta

```python
# Extrai só o texto final
result = resp.json()["outputs"][0]["outputs"][0]["results"]["text"]["text"]
print(result)
```
